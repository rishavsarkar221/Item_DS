```python
import pandas as pd
```

```python
df = pd.read_csv("Practice 2 Sample Data.csv")
df
```

```python
def get_data(array, category, type_of_data):    
    for one_data in array:
        print(f"Finding All The Datas from {type_of_data}: {one_data}\n")
        print(df[df[category] == one_data])
        print('\n\n\n')
```

<h3>To Show Different Types of Items Available

```python
items = []

for each_item in df['Item']:
    if not each_item in items:
        items.append(each_item)
        
print(items)
```

```python
get_data(items, 'Item', 'Item')
```

<div id="region"><h1 style="color: green">Region</h1></div>

```python
regions = []

for each_regions in df['Region']:
    if not each_regions in regions:
        regions.append(each_regions)
regions
```

```python
get_data(regions, 'Region', 'Region')
```

<div id="sorted_data"><h1 style="color: green">Sorted Data</h1>

```python
totals = []

for i in df['Total']:
    totals.append(i)
```

```python
totals.sort(reverse=True)  # Rearranges all the datas in descending order
```

```python
sorted_data = {
    "OrderDate": [],
    "Region": [],
    "Rep": [],
    "Item": [],
    "Units": [],
    "Unit Cost": [],
    "Total": []
}
```

```python
def sorted_data_calc(type_of_data):
    for z in range(len(totals)):
        OrderDate = df['OrderDate'][df['Total'] == totals[z]]
        Region = df['Region'][df['Total'] == totals[z]]
        Rep = df['Rep'][df['Total'] == totals[z]]
        Item = df['Item'][df['Total'] == totals[z]]
        Units = df['Units'][df['Total'] == totals[z]]
        Unit_Cost = df['Unit Cost'][df['Total'] == totals[z]]
        Total = df['Total'][df['Total'] == totals[z]]

        print("**Finding Tabular Range from Total", str(totals[z]) + '**' + '\n')
        
        print("Order Date:\n", str(pd.core.series.Series(OrderDate)) + '\n')
        print("Region:\n",  str(pd.core.series.Series(Region)) + '\n')
        print("Rep:\n",  str(pd.core.series.Series(Rep)) + '\n')
        print("Item:\n",  str(pd.core.series.Series(Item)) + '\n')
        print("Units:\n", str(pd.core.series.Series(Units)) + '\n')
        print("Unit Cost:\n", str(pd.core.series.Series(Unit_Cost)) + '\n')
        print("Total:\n", str(pd.core.series.Series(Total)) + '\n')

        print("\n\n\n\n\n\n")
```
